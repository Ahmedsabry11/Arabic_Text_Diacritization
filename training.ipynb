```python
import matplotlib.pyplot as plt
import numpy as np

from data_preprocessing import DataPreprocessing

test_data_preprocessor = DataPreprocessing()
train_data = test_data_preprocessor.load_text("dataset/undiacritized_test_preprocessed.txt")
# get the length of each sentence
length = []
pad_length = []
count_300 = 0
count_600 = 0
for sentence in train_data:

    if len(sentence) <= 280:
        length.append(len(sentence))
        pad_length.append(280-len(sentence))
    else:
        length.append(280)
        pad_length.append(0)
    
    if len(sentence) > 280:
        count_300 += 1
    if len(sentence) > 600:
        count_600 += 1
print('number of sentences longer than 300: {}'.format(count_300))
print('number of sentences longer than 600: {}'.format(count_600))

# Create a box-and-whisker plot of the training dataset
print(length)
# plt.boxplot(length)
# plt.show()
# Print the number of sentences
print('Total training sentences: {}'.format(len(train_data)))
# Print the longest sentence length
print('Longest training sentence: {}'.format(max(length)))
# Print the shortest sentence length
print('Shortest training sentence: {}'.format(min(length)))
# Print the average sentence length
print('Average training sentence length: {}'.format(np.mean(length)))

# print sum of all characters in padded sentences
print('Total characters in padded sentences: {}'.format(sum(pad_length)))
# print sum of all characters in unpadded sentences
print('Total characters in unpadded sentences: {}'.format(sum(length)))
```
